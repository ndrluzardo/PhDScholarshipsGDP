```python
import statsmodels.formula.api as smf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
bolsas=pd.read_csv('bolsas.csv')
```

```python
model_gdp = smf.ols(formula='bolsa_dollar ~ GDP_dollar', data=bolsas).fit()
print(model_gdp.summary())
```

```python
model_gdp_bigmac = smf.ols(formula='bolsa_dollar ~ GDP_dollar + bigmac_dollar', data=bolsas).fit()
print(model_gdp_bigmac.summary())
```

```python
model_bigmac = smf.ols(formula='bolsa_dollar ~ bigmac_dollar', data=bolsas).fit()
print(model_bigmac.summary())
```

```python
BigMac_GDP_corr = smf.ols(formula='bigmac_dollar ~ GDP_dollar', data=bolsas).fit()
print(BigMac_GDP_corr.summary())
```

```python
# calculate the predicted result for Brazil using model GDP
bolsa_pred_dollar=model_gdp.predict(bolsas.loc[bolsas['country_code']=='BRA','GDP_dollar'])
bolsa_pred_reais=bolsa_pred_dollar*bolsas.loc[bolsas['country_code']=='BRA','dollar_ex']
print('Bolsa mensal prevista em dollar:',float(bolsa_pred_dollar/12))
print('Bolsa mensal prevista em reais:',float(bolsa_pred_reais/12))
```

```python
# calculate the predicted result for Brazil using model GDP+bigmac
bolsa_pred_dollar=model_gdp_bigmac.predict(bolsas.loc[bolsas['country_code']=='BRA',['GDP_dollar','bigmac_dollar']])
bolsa_pred_reais=bolsa_pred_dollar*bolsas.loc[bolsas['country_code']=='BRA','dollar_ex']
print('Bolsa mensal prevista em dollar:',float(bolsa_pred_dollar/12))
print('Bolsa mensal prevista em reais:',float(bolsa_pred_reais/12))
```
